```python
!pip install googlemaps

import googlemaps
import pandas as pd
import numpy as np
import xlrd

# Inicialize o cliente com sua API Key
gmaps = googlemaps.Client(key='COLOQUE SUA CHAVE AQUI')

from google.colab import drive
drive.mount('/content/drive')
```

```python
# Carregar o arquivo Excel em um dataframe
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/PDA_Cursos_Especializacao_Brasil v2.xlsx')
print("DataFrame carregado com sucesso!")
total_linhas = len(df)
```

```python
# Pesquisar um local (Geocoding)
def geocode_address(address):
    geocode_result = gmaps.geocode(address)
    if geocode_result:
        location = geocode_result[0]
        return location['formatted_address'], location['geometry']['location']['lat'], location['geometry']['location']['lng']
    else:
        return None, None, None

# Acha endereço e lat/long e preenche arquivo excel
local_anterior = ""
try:
    #for i in range(0, total_linhas, 1):
    for i in range(0, total_linhas, 1):
        if df.loc[i,'SITUACAO'] == 'Ativo':
          print(f"Processando linha {i} de {total_linhas}")
          local = df.iloc[i, 1] + ',' + df.iloc[i, 2] + ',' + df.iloc[i, 3]
          if local != local_anterior:
            local_anterior = local
            endereco, lat, lng = geocode_address(local)
          df.loc[i, 'ENDERECO'] = endereco
          df.loc[i, 'LATITUDE'] = str(lat)
          df.loc[i, 'LONGITUDE'] = str(lng)

except FileNotFoundError:
    print("Erro: Arquivo Excel não encontrado. Por favor, verifique o caminho e o nome do arquivo.")
except Exception as e:
    print(f"Ocorreu um erro ao carregar o arquivo Excel: {e}")
```
