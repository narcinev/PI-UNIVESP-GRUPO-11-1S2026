```python
#!pip install googlesearch-python
!pip install ddgs
import pandas as pd
import time
#from googlesearch import search
from ddgs import DDGS

from google.colab import files
from google.colab import drive
```

```python
# Carregar o arquivo Excel em um dataframe
drive.mount('/content/drive')
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/CAPES - IES - Brasil (GEOREFERENCIADA) V4.xlsx')
print("DataFrame carregado com sucesso!")
total_linhas = len(df)
print(f"Total de linhas: {total_linhas}")
```

```python
# Pesquisar homepage
def encontrar_homepage(nome_ies):
    try:
        # Criamos uma query específica para melhorar a precisão
        query = f"site oficial da pós graduação da {nome_ies} -site:wikipedia.org -site:facebook.com -site:instagram.com"
        #query = f"site oficial da pós graduação da universidade federal do amazonas -site:wikipedia.org -site:facebook.com -site:instagram.com"
        # Pegamos o primeiro resultado da busca (stop=1)
        # for resultado in search(query, num_results=1):
            #print(resultado)
            #return resultado

        with DDGS() as ddgs:
          resultado = [r for r in ddgs.text(query, max_results=1)]
          print(resultado)
          return resultado
    except Exception as e:
        print(f"Erro ao buscar {nome_ies}: {e}")
        return "Não encontrado"

# Processar a planilha
df2 = df[df['Homepage'] == 'Não encontrado'].copy()
print(len(df['IES - NOME'].unique()), len(df2))
ies_unicas = df2['IES - NOME'].unique()
total_linhas = len(ies_unicas)
print(f"Total de instituições únicas: {total_linhas}")
mapa_urls = {}

print(f"Iniciando busca para {len(ies_unicas)} instituições únicas...")
i = 1
for ies in ies_unicas:
    print(f"({i}/{total_linhas})")
    i += 1
    print(f"Buscando: {ies}")
    mapa_urls[ies] = encontrar_homepage(ies)
    # Pausa de segurança para evitar bloqueio do Google (IP ban)
    time.sleep(10)

#Mapear os resultados de volta para o DataFrame original
df['Homepage'] = df['IES - NOME'].map(mapa_urls)

```

```python
# Gerar a nova versão da planilha
output_name = 'CAPES - IES - Brasil (GEOREFERENCIADA) V41.xlsx'
df.to_excel(output_name)

print(f"\nProcesso concluído! Arquivo '{output_name}' gerado.")

# aixar o arquivo automaticamente
files.download(output_name)
```
